```python
#...............................................Emotion Recognition(Testing)............................................#
import cv2
import time
import numpy as np
from model import FacialExpressionModel
import warnings
warnings.filterwarnings(action = 'ignore')
import datetime
rgb = cv2.VideoCapture(0)
facec = cv2.CascadeClassifier('Dataset/haarcascade_frontalface_default.xml')
font = cv2.FONT_HERSHEY_SIMPLEX

def __get_data__():

    _, fr = rgb.read()
    gray = cv2.cvtColor(fr, cv2.COLOR_BGR2GRAY)
    faces = facec.detectMultiScale(gray, 1.3, 5)
    
    return faces, fr, gray

def start_app(cnn):
    skip_frame = 10
    data = []
    flag = False
    ix = 0
    while True:
        ix += 1
        
        faces, fr, gray_fr = __get_data__()
        for (x, y, w, h) in faces:
            fc = gray_fr[y:y+h, x:x+w]
            
            roi = cv2.resize(fc, (48, 48))
            pred = cnn.predict_emotion(roi[np.newaxis, :, :, np.newaxis])

            cv2.putText(fr, pred, (x, y), font, 1, (255, 255, 0), 2)
            cv2.rectangle(fr,(x,y),(x+w,y+h),(255,0,0),2)
            now = datetime.datetime.now()
            f=open("output.txt",'a')
            f.write(str(now.strftime("%Y-%m-%d %H:%M:%S"))+'-->'+str(pred)+"\n")
            f.close()

        if cv2.waitKey(1) == 27:
            break
        cv2.imshow('Emotion', fr)
    cv2.destroyAllWindows()


if __name__ == '__main__':
    model = FacialExpressionModel("face_model.json", "face_model.h5")
    start_app(model)
```
